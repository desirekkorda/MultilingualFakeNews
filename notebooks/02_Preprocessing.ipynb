```python
# Loading the dataset
import pandas as pd

df = pd.read_csv(
    "/content/drive/MyDrive/MultilingualFakeNews/data/english_clean.csv"
)

print(df.shape)
df.head()
```

```python
# Inspecting text lengths
df["length"] = df["text"].apply(len)

df["length"].describe()
```

```python
import re

# Basic Text Cleaner
def clean_text(text):

    text = str(text)

    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
df["text"] = df["text"].apply(clean_text)
```

```python
# Save the clean dataset
df.to_csv(
    "/content/drive/MyDrive/MultilingualFakeNews/data/english_clean_v2.csv",
    index=False
)
```
