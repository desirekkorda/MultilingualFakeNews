```python
import zipfile

zip_path = "/content/drive/MyDrive/MultilingualFakeNews/data/TALLIP-FakeNews-Dataset.zip"

extract_path = "/content/TALLIP"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted.")
```

```python
#Load first dataset
import pandas as pd

file_path = "/content/TALLIP/TALLIP-FakeNews-Dataset/English/Train/train_English_Data_Complete_FakeNews.txt"

df = pd.read_csv(
    file_path,
    sep="\t"
)

df.head()
```

```python
#Check dataset size
print(df.shape)
```

```python
#Check labels
df["Label"].value_counts()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
# Checking the domains
df["Domain"].value_counts()
```

```python
df.shape
df["Label"].value_counts()
df.isnull().sum()
```

```python
# Create a clean dataset
df = df.dropna(subset=["News", "Label"])

print(df.shape)
```

```python
# Inspect label distribution
df["Label"].value_counts()
```

```python
#Inspect domain distribution
df["Domain"].value_counts()
```

```python
# Creating the training text column
df["Topic"] = df["Topic"].fillna("")

df["text"] = (
    df["Topic"].astype(str)
    + " "
    + df["News"].astype(str)
)
```

```python
df[["text", "Label"]].head()
```

```python
# Convert labels to numbers
label_map = {
    "Fake": 1,
    "Legit": 0
}

df["target"] = df["Label"].map(label_map)

df[["Label", "target"]].head()
```

```python
# Saving the clean dataset
df[["text", "target"]].to_csv(
    "/content/drive/MyDrive/MultilingualFakeNews/data/english_clean.csv",
    index=False
)
```
