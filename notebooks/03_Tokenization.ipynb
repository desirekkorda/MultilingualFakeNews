```python
# Loading the dataset
import pandas as pd

df = pd.read_csv(
    "/content/drive/MyDrive/MultilingualFakeNews/data/english_clean_v2.csv"
)

print(df.shape)
```

```python
# Load XLM-RoBERTa Tokenizer
from transformers import AutoTokenizer

MODEL_NAME = "xlm-roberta-base"

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME
)
```

```python
# Test Tokenization
sample_text = df.iloc[0]["text"]

tokens = tokenizer(
    sample_text,
    truncation=True,
    padding="max_length",
    max_length=256
)

print(len(tokens["input_ids"]))
```

```python
# Inspect Tokens
print(
    tokenizer.convert_ids_to_tokens(
        tokens["input_ids"][:50]
    )
)
```

```python
# Conerting the entire dataset
encodings = tokenizer(
    df["text"].tolist(),
    truncation=True,
    padding=True,
    max_length=256
)
```

```python
# Check
len(encodings["input_ids"])
```

```python
# Additional verification
sample_text = df.iloc[0]["text"]

encoded = tokenizer(
    sample_text,
    truncation=True,
    max_length=256
)

print("Tokens:", len(encoded["input_ids"]))
print(sample_text[:500])
```
